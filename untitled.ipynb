```python
#importing all the necessary packages

import os
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the sample data
data = pd.read_csv('sample_data.txt', sep = ', ', engine = 'python')

#stripping '[' and ']' from the timestamp columns
data['Timestamp'] = data['Timestamp'].map(lambda x: x.lstrip('[').rstrip(']'))

#converting timestamp to datetime format.
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

#mapping to map True to 1 and False to 0. This way we can use make plots and model.
mapping = {True: 1, False: 0}
data = data.applymap(lambda s: mapping.get(s) if s in mapping else s)
```

## Data Visualization and Exploration

```python
### Helper function to plot the individual signals

def plot_signal(data, signal, color = None):
    
    '''
    data -- pandas dataframe to be used
    signal -- column name for the signal to be plotted
    color -- default = None, color for the line plot
    '''
    
    fig = plt.figure(figsize=(21, 5));
    ax = fig.add_subplot(111)
    ax.step(data['Timestamp'],data[signal],color=color,label=signal)
    ax.legend(fontsize = 12, loc = 'best')
    plt.xlabel('Timestamp', fontsize = 15);
    plt.xticks(fontsize=15, rotation=45)
    plt.yticks(fontsize=15)
    plt.ylabel(signal, fontsize = 15);
    plt.show()
```

```python
###plotting the 4 streams
plot_signal(data,'Stream1', 'b')
plot_signal(data,'Stream2', 'r')
plot_signal(data,'Stream3', 'y')
plot_signal(data,'Stream4', 'g')
```

```python
plot_signal(data,'Overall')
plot_signal(data,'Truth')
```

```python
fig = plt.figure(figsize=(20, 5));
ax = fig.add_subplot(111)
ax.step(data['Timestamp'],data['Overall'],color='r',label='Overall')
ax.step(data['Timestamp'],data['Truth'],color='b',label='Truth')
ax.legend(fontsize = 12, loc = 'best')
plt.xlabel('Timestamp', fontsize = 15);
plt.xticks(fontsize=15, rotation=45);
plt.yticks(fontsize=15);
```

#### A lag is observed between the truth and overall. Initially at the start, the truth goes to 1 and overall stays 0 (all streams are 0). After that we observe the opposite nature of lag, truth goes to 0 and overall stays 1. 

```python
fig = plt.figure(figsize=(20, 5));
ax = fig.add_subplot(111)
ax.step(data['Timestamp'], data.Stream1 | data.Stream2 | data.Stream4,color='r',label='OR')
ax.legend(fontsize = 12, loc = 'best')
plt.xlabel('Timestamp', fontsize = 15);
plt.xticks(fontsize=15, rotation=45);
plt.yticks(fontsize=15);
```

###### The above plot reveals a major thing -- the overall signal looks like a simple OR of all 4 streams with a lag from previous timesteps.

```python
fig = plt.figure(figsize=(20, 5));
ax = fig.add_subplot(111)
plt.plot((data['Timestamp'] - data['Timestamp'].shift()).dt.total_seconds())
plt.title('Time Interval Between Observations - Frequency', fontsize = 15);
plt.yticks(fontsize = 15);
```

### Note
Time intervals are not constant. We need to fill the missing values. Most of the time series modeling techniques require the time intervals to be uniform. The streams are not that volatile. Hence we can fill in the missing values.
One way could be is to sample the timestamps into constant bins and use forward fill i.e. use the last value forward. 

```python
mod_data = data.set_index('Timestamp').resample('100ms').ffill().reset_index().dropna()
```

```python
print('Length of data before filling missing values', len(data))
print('Length of data after filling missing values', len(mod_data))
```
